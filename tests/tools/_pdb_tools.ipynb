```python
all_pdb_files = """
101M/101M.pdb
1A0Z/1A0Z.pdb
1A6B/1A6B.pdb
1ATN/1ATN_1w.pdb
1ATN/1ATN_2w.pdb
1ATN/1ATN_3w.pdb
1ATN/1ATN_4w.pdb
1CRN/1CRN.pdb
1ak4/1ak4.pdb
2g98/pdb2g98.pdb
3C8P/3C8P.pdb
3MRC/3MRC.pdb
9api/9api.pdb
"""

all_pdb_files = all_pdb_files.splitlines()[1:]
base_folder = "/home/dbodor/git/DeepRank/DeepRank2/tests/data/pdb//"

all_pdb_files
```

```python
tell_me = {}

for file in all_pdb_files:
    with open(base_folder + file, "r") as f:
        records = f.read().splitlines()

    for i, line in enumerate(records):
        if line.startswith(("ATOM", "HETATOM")):
            ins_code = slice(26, 27)
            if line[ins_code].strip():
                inscode: list = tell_me.setdefault("inscode", [])
                inscode.append((file, i, line[ins_code]))

            alt_loc = slice(16, 17)
            if line[alt_loc].strip():
                altloc: list = tell_me.setdefault("altloc", [])
                altloc.append((file, i, line[alt_loc]))

            resname = slice(17, 20)
            if line[resname] == "HOH":
                water: list = tell_me.setdefault("water", [])
                water.append((file, i, line[resname]))

tell_me["altloc"]
```
